```python
from telethon import TelegramClient
import pandas as pd
import logging
import os
import asyncio
import nest_asyncio

nest_asyncio.apply()

# Set up logging
logging.basicConfig(level=logging.INFO)
logger = logging.getLogger(__name__)

# Telegram API 
api_id = 26205886  
api_hash = '78566b82e3bf847041fe85a9d53b2d97'
phone_number = '+251939836983'

client = TelegramClient('Telegeram channels', api_id, api_hash)

# List of channels to scrape
channels = [
    'https://t.me/DoctorsET',
    'https://t.me/CheMed123',  
    'https://t.me/lobelia4cosmetics',
    'https://t.me/yetenaweg',
    'https://t.me/EAHCI'
]

# Directory for temporary storage of images
os.makedirs('scraped_images', exist_ok=True)

# Function to scrape messages and images from a channel
async def scrape_channel(channel):
    try:
        
        entity = await client.get_entity(channel)
        data = []
        
        async for message in client.iter_messages(entity):
            message_data = {
                'date': message.date,
                'sender': message.sender_id,
                'content': message.text
            }
            if message.media:
                if hasattr(message.media, 'photo'):
                    # Download the image
                    file_path = await client.download_media(message.media, file='scraped_images/')
                    message_data['image'] = file_path

            data.append(message_data)

        # Convert to DataFrame and save or process further
        df = pd.DataFrame(data)
        logger.info(f"Scraped {len(df)} messages from {channel}")

        # Save the DataFrame to CSV
        df.to_csv(f"{channel.split('/')[-1]}.csv", index=False)

    except Exception as e:
        logger.error(f"Error scraping {channel}: {str(e)}")

# Main function to run the scraping
async def main():
    async with client:
        logger.info("Client Created. Starting to scrape channels...")
        for channel in channels:
            await scrape_channel(channel)

# Execute the scraper
if __name__ == "__main__":
    asyncio.run(main())
```
